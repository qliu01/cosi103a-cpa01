# Diamonds Dataset

## Dataset Description

This a dataset containing information about diamond prices, with various features like number of carats, color, cut, clarity, and diamond dimensions available.

## URL

The dataset is from Vincent Arel-Bundock, a political scientist working at University of Montreal. The URL is https://vincentarelbundock.github.io/Rdatasets/csv/ggplot2/diamonds.csv and the dataset can be downloaded from this link.

## Two Interesting Questions

Question 1. Fixing the carat dimension, what is the worst diamond clarity (measuring against price)?

Question 2. Fixing the depth dimension, what is the color hierarchy (measuring against price), amongst the diamond colors D, G, and J?

```python
carat_prices = pd.pivot_table(diamonds_df, values='price', index=["carat"], columns=['clarity'])
```

```python
carat_prices[['IF']].plot(figsize=(15,10))
plt.grid()
plt.title("Carats vs Price for Diamond Clarity Level of IF",fontsize=32)
```

```python
carat_prices[['IF','VS1','I1']].plot(figsize=(15,10))
plt.grid()
plt.title("Carats vs Price for Three Diamond Clarity Levels: IF, VS1, and I1",fontsize=32)
```

```python
fac_carat = diamonds_df.groupby(pd.cut(diamonds_df["carat"], np.arange(0, 5, 0.5)))
fac_carat['price'].agg(np.mean)
```

```python
carat_prices.plot(figsize=(15,10))
plt.grid()
plt.title("Carats vs Price for All Clarity Levels",fontsize=32)
```

Analysis for question 1: According to this plot, it's clear that fixing the carat dimension, the "I1" clarity has the worst price, which is consistent with actual diamond ratings (I1 is the lowest).

```python
x_prices_color = pd.pivot_table(diamonds_df, values='price', index=["z"], columns=['color'])
x_prices_color[['D','G','J']].plot(figsize=(15,10))
plt.grid()
plt.title("Depth vs Price for Diamond Colors of D, G, and J",fontsize=32)
```

Analysis for question 2: According to this plot, especially at the lower depth values, the it's clear that fixing the carat dimension, D has the best price, G has the second best price, and J has the worst price. At higher carat values, the hierarchy starts to become a bit more muddled although the hierarchy is generally true. This indicates that at higher carat values, color may matter less and other factors (perhaps clarity start to play a bigger role.
